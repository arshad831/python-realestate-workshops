```python
import pandas as pd

# Convert 'contract_start_date' to datetime format
df['contract_start_date'] = pd.to_datetime(df['contract_start_date'], errors='coerce')

# Check if conversion was successful
if df['contract_start_date'].isnull().sum() > 0:
    print("There were errors in the date conversion.")

# Extract the month and year from 'contract_start_date'
df['year_month'] = df['contract_start_date'].dt.to_period('M')

# Filter contracts between July 2020 and July 2023
start_date = pd.Period('2020-07', freq='M')
end_date = pd.Period('2023-07', freq='M')
df_filtered = df.loc[(df['year_month'] >= start_date) & (df['year_month'] <= end_date)]

# Count the number of open contracts per month
open_contracts = df_filtered['year_month'].value_counts().sort_index()

# Print the result
print(open_contracts)
```
